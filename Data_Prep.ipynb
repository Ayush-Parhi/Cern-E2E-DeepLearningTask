# Importing essential Dependencies

```python
import h5py
import pandas as pd
import numpy as np
import random 
```

# Reading Photon Dataset

```python
with h5py.File("/content/drive/My Drive/collab notebooks/SinglePhotonPt50_IMGCROPS_n249k_RHv1.hdf5", 'r') as f:
    print(f.keys())
    a_group_key = list(f.keys())[0]
    b_group_key = list(f.keys())[1]
    pdata = list(f[a_group_key][:]) # pdata is a list of data/features of photons  
    plabel = list(f[b_group_key][:])# plabel is the list of label which is given to photon for categorization
```

```python
plabel
```

# Reading Electron Dataset

```python
with h5py.File("/content/drive/My Drive/collab notebooks/SingleElectronPt50_IMGCROPS_n249k_RHv1.hdf5", 'r') as f2:
    print(f2.keys())
    a_group_key2 = list(f2.keys())[0]
    b_group_key2 = list(f2.keys())[1]
    edata = list(f2[a_group_key2][:]) # edata is a list of data/features of electrons
    elabel = list(f2[b_group_key2][:]) # elabel is a list of label which is given to electrons for categorization
```

# Training Data Preparation

```python
pdata = np.array(pdata)
```

```python
pdata.shape
```

```python
pdata = pdata.reshape(249000,32*32*2)
```

```python
df1 = pd.DataFrame(pdata)
df1
```

```python
df1.describe()
```

```python
edata = np.array(edata)
edata.reshape(249000,32*32*2)
edata.shape
```

```python
edata = edata.reshape(249000, 32*32*2)
```

```python
df2 = pd.DataFrame(edata)
df2
```

```python
df2.describe()
```

```python
frames = [df1,df2]
df = pd.concat(frames, ignore_index = True)
df
```

```python
df.describe()
```

```python
labels = plabel + elabel
labels
```

```python
df["Labels"] = labels
df
```

```python
df = df.sample(frac=1).reset_index(drop=True)
```

```python
df.describe()
```

# Saving the Dataframe
